```
import random as rd
from openpyxl import Workbook
from openpyxl import Workbook, load_workbook
from openpyxl.drawing.image import Image
from openpyxl.utils.dataframe import dataframe_to_rows

def machine(x,y):
  #list to append values to excel
  lista_insert = []

  ## variables ##
  pocket = x
  bet_value = y
  
  #wins and losses counters
  win_t = lose_t = 0
  #--
  number_of_bets =0
  profit = 0
  #--
  money = pocket

  # loop that creates the data
  while pocket >= bet_value:
    #empty list that receives 3 random values
    lista_num = []
    #loop append 3 random values to the empty list
    for i in range(0,3):
      lista_num.append(rd.randrange(0,4))
    #compares the lista_num values for the win condition and iterates couters
    if lista_num[0] == lista_num[1] == lista_num[2]:
      #adds win money to player
      profit += bet_value * 10  #win multiplier
      win_t += 1
        
    else:
      pocket -= bet_value
      lose_t +=1

    number_of_bets +=1
  
  #calculate player profit
  player_profit = profit - money

  #calculate machine profit
  money = money - pocket - profit
  
  #insertes variables into the excel append list
  lista_insert.extend([money,bet_value,win_t,lose_t,number_of_bets,player_profit])
  print(lista_insert)

  ## excel append section ##

  #path to the excel file
  excel_file = '/content/drive/My Drive/slotMachDatasets/slotMachine.xlsx'

  #load excel file
  wb = load_workbook(excel_file)

  #declares the active sheet were to append
  f1 = wb.active
  #appends values to declared sheet
  f1.append(lista_insert)

  #saves changes made to the excel file
  wb.save(filename=excel_file)

#counter for the loop
entry_count = 0

#loop to determine how many entries are added to the excel file
#How many times to run the function machine(x,y)
loop_runs = 10
#loop_runs = int(input("Number of entries to create:"))

while entry_count < loop_runs:
  #value of money inserte into the machine
  x = rd.randrange(10,2000) 
  #value of the bet
  y = rd.randrange(1,200)

  #money value is higher then bet value
  if x > y:
    #function call
    machine(x,y)
    #counter iteration
    entry_count += 1

#Values are appended in the excel file, it may append in the end of the document
# even if its empty

# Win probability

#number_of_winning_combinations = 4
#number_caracters_per_value = 4
#number_of_entries = 3
#win_prob = number_of_winning_combinations / (number_caracters_per_value * number_of_entries)

# win_prob = 0,33
```
